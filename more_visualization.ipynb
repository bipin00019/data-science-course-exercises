```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig, ax = plt.subplots() # creates a fig and axes

x = [1, 2, 3]
y = [4, 5, 6]

ax.plot(x,y)
ax.set_title("OOP Interface")
ax.set_xlabel("X-axis")
ax.set_ylabel("Y_axis")

plt.show()
```

```python
x = np.arange(-10,11)

fig, axes = plt.subplots(figsize=(12,6))
axes.plot(
    x,(x**2), color='red', linewidth = 3, marker = 'o', markersize = 8, label = 'x^2'
)

axes.plot(
    x, -1 * (x**2), 'b--', label = '-x^2'
)

axes.set_title("My plot")
axes.set_xlabel("x")
axes.set_ylabel("x squared")
axes.legend()

plt.show()
```

```python
fig, axes = plt.subplots(figsize = (12,6))

x = np.arange(0,11)
axes.plot(x, x+0, linestyle = 'solid')
axes.plot(x, x+1, linestyle = 'dashed')
axes.plot(x, x+2, linestyle = 'dashdot')
axes.plot(x, x+3, linestyle = 'dotted')
```

```python
fig, axes = plt.subplots(figsize = (12,6))
x = np.arange(0,11)

axes.plot(x, x+0,'-og', label="solid green")
axes.plot(x, x-1,'-ob', label="dashdot blue")
axes.legend()
plt.show()
```


**Other types of plots**

Figures and subfigures

When we call the subplots() function we get a tuple containing a Figure and a axes element.

```python
plot_objects = plt.subplots() # here fig,axes as tuple is stored in the same variable
fig, axes = plot_objects

axes.plot([1,2,3],[1,2,3])
```

We can also define how many elements we want inside our figure. To do that we can set the nrows and ncols params.

```python
plot_objects = plt.subplots(nrows = 2, ncols = 2, figsize = (14,6))

fig, ((ax1, ax2), (ax3, ax4)) = plot_objects
plot_objects
```

```python

ax1.plot(np.random.randn(50), c='red', linestyle='--')
ax2.plot(np.random.randn(50), c='green', linestyle=':')
ax3.plot(np.random.randn(50), c='blue', marker='o', linewidth=3.0)
ax4.plot(np.random.randn(50), c='yellow')

fig
```

The subplot2grid command

There is another way to make subplots using a grid-like format:

```python

plt.figure(figsize=(14, 6))

ax1 = plt.subplot2grid((3,3), (0,0), colspan=3)
ax2 = plt.subplot2grid((3,3), (1,0), colspan=2)
ax3 = plt.subplot2grid((3,3), (1,2), rowspan=2)
ax4 = plt.subplot2grid((3,3), (2,0))
ax5 = plt.subplot2grid((3,3), (2,1))
```

Scatter plot

```python
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
area = np.pi * (20 *np.random.rand(N))**2 # area of cirle and radius is multipied by 20 just to make it more visible
```

```python
plt.figure(figsize = (12, 6))
plt.scatter(x, y, s = area, c = colors, alpha=0.5 , cmap="Spectral")
plt.colorbar()
plt.show()
```

```python
fig = plt.figure(figsize = (14, 6))

ax1 = fig.add_subplot(1,2,1) # add_subplot(no of rows, no of column and index of column)
plt.scatter(x, y, s = area, c =colors, cmap = 'Pastel1')
plt.colorbar()

ax1 = fig.add_subplot(1,2,2) # add_subplot(no of rows, no of column and index of column)
plt.scatter(x, y, s = area, c =colors, cmap = 'Pastel2')
plt.colorbar()

plt.show()
```

```python


# Fixed positions (x, y)
x = [1, 2, 3, 4, 5]
y = [10, 20, 10, 20, 10]

# Fixed sizes for each point
area = [100, 200, 300, 400, 500]  # marker sizes

# Colors array - notice some repeated values
colors = [10, 20, 10, 40, 20]  # numeric values mapped to colors

plt.figure(figsize=(8, 5))

# Scatter plot
plt.scatter(x, y, s=area, c=colors, cmap='Spectral', alpha=0.7)

# Show colorbar with the color mapping of 'colors'
plt.colorbar(label='Color value')

plt.xlabel('X position')
plt.ylabel('Y position')
plt.title('Scatter plot with color mapping independent of position and size')

plt.show()
```

Histogram

```python
values = np.random.randn(1000)
plt.figure(figsize = (12, 6))
plt.hist(values, bins = 100, alpha = 0.7, color='steelblue', edgecolor = 'red')
plt.xlim(xmin=-5, xmax= 5)
plt.show()
```

Bar plots

```python
Y = np.random.rand(1,5)[0] # creates 2d numpy array with row 1 and column 5, and extracts the first row
Y2 = np.random.rand(1,5)[0]

plt.figure(figsize=(12,4))
plt.bar(np.arange(len(Y)), Y)

plt.show()
```


Also can be stacked bars, and add a legend to the plot:

```python
Y = np.random.rand(1,5)[0] # creates 2d numpy array with row 1 and column 5, and extracts the first row
Y2 = np.random.rand(1,5)[0]

plt.figure(figsize=(12,4))
plt.bar(np.arange(len(Y)), Y, label = "label Y")
plt.bar(np.arange(len(Y2)), Y2,bottom=Y, label = "label Y2") # bottom sets the starting of Y2 at the endpoint of Y
plt.legend()
plt.show()
```

Boxplots and outlier detection

```python
plt.figure(figsize=(12,4))
plt.hist(values)
plt.show()
```
