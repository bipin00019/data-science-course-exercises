How do you plot a simple line graph using Matplotlib?

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(1,11)
y = np.arange(11,21)

plt.plot(x,y, marker = 'o')

plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Simple line graph")

plt.show()

```

What is the difference between plt.plot() and plt.scatter()?

```python
# Sample data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.figure(figsize=(10,5))

# Left side
plt.subplot(1,2,1) #(rows, columns, index)
plt.plot(x,y, marker ='o', color='b',label="line plot")
plt.title("plt.plot() Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()

# Right side
plt.subplot(1,2,2)
plt.scatter(x,y, color = 'r', label="Scatter plot")
plt.title("plt.scatter() Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()

# Adjust layout
plt.tight_layout()
plt.show()

# Differences
# plt.plot() -> when we care about the relationship
# plt.scatter() -> when we care individual points
```

How do you change the line color and style in a line plot?

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(1,11)
y = np.arange(11,21)

plt.plot(x,y, linestyle = '--', color='blue',marker = 'o', linewidth = 2, label = "Dashed blue line")

# Colors: 'red', 'blue', 'green', 'black', or short codes: 'r', 'b', 'g', 'k'

#Line styles:
# '-' → solid line (default)
# '--' → dashed line
# '-.' → dash-dot line
# ':' → dotted line


plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Simple line graph")
plt.legend()

plt.show()

```

What does the alpha parameter control?

```python
# alpha parameter controls the transparency

# Generate random data
x = np.random.randn(500)
y = np.random.randn(500)

# Scatter plot with transparency
plt.scatter(x, y, color='green', alpha=0.4, label='Points')

plt.title("Scatter Plot with Transparency")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.show()
```

How do you plot multiple subplots in the same figure?

```python
fig, axs = plt.subplots(2, 1, figsize=(6,8))  # 2 rows, 1 column

x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

axs[0].plot(x, y1, color='blue')
axs[0].set_title("sin(x)")

axs[1].plot(x, y2, color='red')
axs[1].set_title("cos(x)")

plt.tight_layout()
plt.show()
```

What is the difference between plt.bar() and plt.barh()?

```python
import matplotlib.pyplot as plt

# Create figure with 2 subplots (2 rows, 1 column)
fig, axs = plt.subplots(2, 1, figsize=(6, 8))

categories = ['A', 'B', 'C']
values = [5, 7, 3]

# Vertical bar chart in the first subplot
axs[0].bar(categories, values, color='blue')
axs[0].set_title("Vertical Bar Chart")

# Horizontal bar chart in the second subplot
axs[1].barh(categories, values, color='green')
axs[1].set_title("Horizontal Bar Chart")

# Adjust layout and display
plt.tight_layout()
plt.show()
```

How do you save a Matplotlib figure as an image?

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.plot(x, y, marker='o', color='blue')
plt.title("Line Plot Example")

# Save the figure as a PNG image
plt.savefig("line_plot.png")
plt.show()
```

combine two scatter plots in a single figure

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate random data
x1 = np.random.randn(100)
y1 = np.random.randn(100)

x2 = np.random.randn(100) + 2
y2 = np.random.randn(100) + 2

# Create a figure with 1 row and 2 columns
fig, axs = plt.subplots(1, 2, figsize=(12, 5))

# First scatter plot
axs[0].scatter(x1, y1, color='blue', alpha=0.6)
axs[0].set_title("Scatter Plot 1")
axs[0].set_xlabel("X-axis")
axs[0].set_ylabel("Y-axis")

# Second scatter plot
axs[1].scatter(x2, y2, color='red', alpha=0.6)
axs[1].set_title("Scatter Plot 2")
axs[1].set_xlabel("X-axis")
axs[1].set_ylabel("Y-axis")

# Adjust layout and display
plt.tight_layout()
plt.show()
```
